```python
Final_outputs = []
out_ind = 0
ind = 0
j = 0
lst_start = 0
while ind < length:
    if j < len(index_lst) and ind == index_lst[j]:
        lst_start += 37
        j += 1
    
    row = in_df.iloc[ind]
    
    imagenet = []
    coco = []
    sun = []
    
    detrend_in = []
    for pos in range(0,37):
        name = "img" + str(pos+1)
        detrend_in.append(float(row[name]))
        
    detrend_out = signal.detrend(np.array(detrend_in))
    
    coco_pres = False
    imagenet_pres = False
    sun_pres = False
    
    
    for pos in range(0,37):
        out_ind = lst_start + pos
        if out_num[out_ind] == 0:
            coco.append(detrend_out[pos])
            coco_pres = True
        
        elif out_num[out_ind] == 1:
            imagenet.append(detrend_out[pos])
            imagenet_pres = True
            
        elif out_num[out_ind] == 2:
            sun.append(detrend_out[pos])
            sun_pres = True
    
    if coco_pres == True:
        Final_outputs.append(0)
        len_c = 37 - len(coco)
        for i in range(len_c):
            coco.append(0)
        
        data_to_append = {}
        for i in range(len(comp_df.columns)):
            data_to_append[comp_df.columns[i]] = coco[i]

        comp_df = comp_df.append(data_to_append, ignore_index = True)
        
    if imagenet_pres == True:
        Final_outputs.append(1)
        len_i = 37 - len(imagenet)
        for i in range(len_i):
            imagenet.append(0)
            
        data_to_append = {}
        for i in range(len(comp_df.columns)):
            data_to_append[comp_df.columns[i]] = imagenet[i]

        comp_df = comp_df.append(data_to_append, ignore_index = True)
        
    if sun_pres == True:
        Final_outputs.append(2)
        len_s = 37 - len(sun)
        for i in range(len_s):
            sun.append(0)
            
        data_to_append = {}
        for i in range(len(comp_df.columns)):
            data_to_append[comp_df.columns[i]] = sun[i]

        comp_df = comp_df.append(data_to_append, ignore_index = True)
            
    
    ind += 1
    
comp_df
```
