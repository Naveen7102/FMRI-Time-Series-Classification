```python
index_lst = []
for i in range(len(clust_lst)):
    if np.isnan(clust_lst[i]) == False:
        index_lst.append(i)

print(len(index_lst))
print(len(index_lst[1:]))
index_lst = index_lst[1:]
print(len(index_lst))
```

```python
ind = 0
j = 0
lst_start = 0
while ind < 3477:
    if j < len(index_lst) and ind == index_lst[j]:
        lst_start += 37
        j += 1
    
    row = in_df.iloc[ind]
    voxel = str(row['vx']) + " " + str(row['vy']) + " " + str(row['vz'])
    
    detrend_in = []
    for pos in range(0,37):
        name = "img" + str(pos+1)
        detrend_in.append(float(row[name]))
    
    detrend_out = signal.detrend(np.array(detrend_in))
    
    for pos in range(0, 37):
        comp_df[voxel][lst_start + pos] = detrend_out[pos]
    
    ind += 1
```
