```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
r = 0
g = 1
b = 2
```

```python
x1_true = [g,g,g,r,g,b,b,r,r,b,g,r,r,g,r,b,r,b,g,r,g,r,g,g,g,r,b,b,g,r,g,r,g,r,g,g,r]
x1_pred = [g,g,g,g,g,r,r,r,r,b,g,r,r,g,r,b,r,g,r,r,g,r,g,g,g,g,b,b,g,r,r,r,g,r,b,g,r]
```

```python
x2_true = [g,b,r,g,b,g,g,r,b,r,r,r,r,r,b,g,b,g,b,r,r,g,b,g,r,g,g,b,r,r,g,g,r,r,g,r,r]
x2_pred = [g,b,r,r,b,g,r,r,g,r,r,r,r,g,b,g,r,g,g,g,r,g,b,g,r,g,g,b,r,r,g,g,r,r,r,r,r]
```

```python
def plot(x_true,x_pred):
    
    col_true = []
    for i in range(37):
        if(x_true[i] == 0):
            col_true.append('r')
        elif x_true[i] == 1:
            col_true.append('g')
        else:
            col_true.append('b')
    
    col_pred = []
    for i in range(37):
        if x_pred[i] == 0:
            col_pred.append('r')
        elif x_pred[i] == 1:
            col_pred.append('g')
        else:
            col_pred.append('b')
            
    x_val = [i+1 for i in range(37)]
    y_val = [1 for i in range(37)]
    
    plt.bar(x_val, y_val, color = col_true,width = 1)
    plt.bar(x_val, [-i for i in y_val], color = col_pred,width = 1)
    plt.plot([i for i in range(1,38)],[0 for i in range(1,38)], color = 'black')
    frame1 = plt.gca()
    frame1.axes.get_yaxis().set_ticks([0])
    frame1.axes.tick_params(axis='x', colors='black')
    frame1.axes.set_xlabel('time')
#     plt.text(-3, 0.3, "Actual Label",rotation=90)
#     plt.text(-3,-0.9,"Predicted Label",rotation=90)
    plt.text(-10, 0.5, "Actual Label")
    plt.text(-12,-0.5,"Predicted Label")
    red_patch = mpatches.Patch(color='red', label='COCO')
    green_patch = mpatches.Patch(color='green', label='ImageNet')
    blue_patch = mpatches.Patch(color='blue', label='SUN')
    plt.legend(handles=[red_patch,green_patch,blue_patch],bbox_to_anchor=(1.01, 1), loc='upper left')

    plt.show()
```

```python
plot(x1_true,x1_pred)
```
