```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
```

```python
input_data = pd.read_csv("/kaggle/input/time-series-data/Final_inputs_2.csv")
output_data = pd.read_csv("/kaggle/input/time-series-data/Final_outputs_2.csv")
```

```python
temp_coco = input_data.iloc[3].tolist()
temp_image = input_data.iloc[4].tolist()
temp_sun = input_data.iloc[5].tolist()
```

```python
def clean(lst):
    new_lst = []
    for i in lst:
        if i != 0:
            new_lst.append(i)
            
    return new_lst
            
fin_coco = clean(temp_coco)
fin_image = clean(temp_image)
fin_sun = clean(temp_sun)
```
